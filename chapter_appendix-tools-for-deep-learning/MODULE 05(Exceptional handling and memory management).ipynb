## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

##**1. Difference between interpreted and compiled languages**

**Answer:** An interpreted language is executed line by line by an interpreter, which translates and runs the code directly. A compiled language is first translated into machine code by a compiler, and then the entire program is executed.

##**2. Exception handling in Python**

**Answer:**
Exception handling is a mechanism in Python used to manage errors that occur during the execution of a program. It allows a program to continue running even if an error occurs, preventing it from crashing.

##**3. Purpose of the finally block**
**Answer:** The finally block in exception handling is used to define code that will always be executed, regardless of whether an exception occurred in the try block or not. This is useful for cleanup actions, such as closing files or releasing resources.

## **4. Logging in Python**
**Answer:** Logging in Python is the process of recording events that occur while a program is running. It provides a way to track the program's behavior and diagnose problems.

## **5. Significance of the \_del\_ method**
**Answer:** The \_del\_ method, also known as the destructor, is a special method in Python that is called when an object is about to be destroyed. It is used to perform cleanup actions, such as closing connections or releasing memory, before the object is garbage collected.

##**6. Difference between import and from...import**
**Answer:** The import statement imports an entire module, and you must use the module name as a prefix to access its contents (e.g., math.pi). The from...import statement imports specific attributes or functions directly from a module, allowing you to use them without the module name prefix (e.g., pi).

##**7. How to handle multiple exceptions**
**Answer:** You can handle multiple exceptions in Python by using multiple except blocks, each designed to handle a specific type of exception. You can also handle several exceptions with a single except block by providing a tuple of exception types.
Example:
Python

try:

    # code that might raise an exception
except (ValueError, TypeError):

    # handle both ValueError and TypeError

##**8. Purpose of the with statement**
**Answer:** The with statement is used to simplify resource management, particularly for file handling. It ensures that a resource, such as a file, is properly closed after its use, even if an error occurs.

##**9. Difference between multithreading and multiprocessing**
**Answer:** Multithreading involves multiple threads within a single process, sharing the same memory space. It is suitable for I/O-bound tasks. Multiprocessing involves multiple processes, each with its own memory space. It is suitable for CPU-bound tasks.

##**10. Advantages of using logging**
**Answer:** The advantages of using logging include:
Debugging: It helps in tracking the flow of the program and identifying bugs.
Monitoring: It provides a record of events, which is useful for monitoring the program's behavior in a production environment.

Security: It can be used to log security-related events, such as login attempts.

##**11. Memory management in Python**
**Answer:** Memory management in Python is the process of allocating and de-allocating memory for objects. Python uses a private heap for memory allocation and a garbage collector to automatically reclaim memory that is no longer in use.

##**12. Basic steps involved in exception handling**
**Answer:** The basic steps are:

try block: The code that might raise an exception is placed here.

except block: This block catches and handles specific exceptions.

else block (optional): The code in this block is executed if no exceptions occur in the try block.

finally block (optional): The code in this block is always executed, regardless of whether an exception occurred.

##**13. Why memory management is important**
**Answer:** Memory management is important to ensure that a program uses memory efficiently. Poor memory management can lead to memory leaks, where a program consumes more and more memory, eventually causing it to slow down or crash.

##**14. Role of try and except**
**Answer:** The try block contains the code that is being monitored for exceptions. If an exception occurs, the execution of the try block is stopped, and the program jumps to the except block. The except block contains the code to handle the exception, allowing the program to gracefully recover from the error.

##**15. How Python's garbage collection system works**
**Answer:** Python's garbage collection system primarily uses a reference counting mechanism. It keeps track of the number of references to an object. When the reference count drops to zero, the object is de-allocated. It also has a generational garbage collector to handle reference cycles.

##**16. Purpose of the else block**
**Answer:** The else block in exception handling is executed only if the code in the try block runs without raising any exceptions. It is often used for code that should only be executed if the try block was successful.
##**17. Common logging levels**
**Answer:** The common logging levels in Python, in order of severity, are:

DEBUG: *Detailed information, typically of interest only when diagnosing problems.*

INFO: *Confirmation that things are working as expected.*

WARNING: *An indication that something unexpected happened, or a problem is brewing.*

ERROR: *A more serious problem, but the program is still running.*

CRITICAL: *A serious error, indicating the program may be unable to continue running.*

##**18. Difference between os.fork() and multiprocessing**
**Answer:** os.fork() is a low-level function that creates a new process by duplicating the current one. It is available only on Unix-like systems. The multiprocessing module is a high-level, cross-platform module that provides an API for creating and managing processes, making it easier to write parallel programs.

##**19. Importance of closing a file**
**Answer:** It is important to close a file after you are done with it to free up system resources. If a file is not closed, it can lead to resource leaks and may prevent other programs from accessing the file.

##**20. Difference between file.read() and file.readline()**
**Answer:** file.read() reads the entire content of a file into a single string. file.readline() reads a single line from the file at a time. file.read() is useful for small files, while file.readline() is more memory-efficient for large files.

##**21. What the logging module is used for**
**Answer:** The logging module is used to provide a flexible framework for logging events from applications and libraries. It allows developers to record messages at different severity levels and direct them to various destinations, such as the console or a file.

##**22. What the os module is used for in file handling**
**Answer:** The os module provides a way of using operating system-dependent functionality. In file handling, it is used for tasks like creating and removing directories, getting file metadata, changing file permissions, and navigating the file system.

##**23. Challenges associated with memory management**
**Answer:** Challenges include:

Memory leaks: Objects are not properly de-allocated, leading to a gradual increase in memory usage.

Fragmentation: Memory becomes divided into small, non-contiguous blocks, making it difficult to allocate larger objects.

Garbage collection overhead: The process of garbage collection can sometimes consume significant CPU time, affecting performance.

##**24. How to raise an exception manually**
**Answer:** You can raise an exception manually using the raise statement, followed by the exception object.
Example:
Python

raise ValueError("Invalid value provided.")

##**25. Why it is important to use multithreading**
**Answer:** Multithreading is important for applications that have I/O-bound tasks, such as network communication or file operations. It allows a program to perform other tasks while waiting for these slow operations to complete, improving the responsiveness and overall performance of the application
